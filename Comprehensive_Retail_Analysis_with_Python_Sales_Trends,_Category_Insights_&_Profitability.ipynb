```python
import pandas as pd #pandas (pd): Data manipulation ke liye use hota hai, jaise CSV file load karna aur process karna.

import plotly.express as px #plotly.express (px): Data visualization library jo easy aur quick plots banata hai.
import plotly.graph_objects as go #plotly.graph_objects (go): Advanced and customizable graphs banane ke liye
import plotly.io as pio   #plotly.io (pio): Graph templates ko customize karne ke liye
import plotly.colors as colors
pio.templates.default = "plotly_white"  #pio.templates.default = "plotly_white": Default theme white rakha gaya hai graphs ke liye
```

```python
data = pd.read_csv("/content/drive/MyDrive/Portofolio/Sample - Superstore.csv", encoding='latin-1')
#encoding='latin-1': Special characters ko properly read karne ke liye encoding use hui hai
data.head()
```

# Converting Date Columns

```python
data['Order Date'] = pd.to_datetime(data['Order Date'])
data['Ship Date'] = pd.to_datetime(data['Ship Date'])
#Date Conversion: Order Date aur Ship Date columns ko datetime format me convert kiya gaya hai for date-based analysis.
```

# Adding New Date-Based Columns
The dataset has an order date column. We can use this column to create new columns like order month, order year, and order day, which will be very valuable for sales and profit analysis according to time periods. So let’s add these columns:

```python
data['Order Month'] = data['Order Date'].dt.month
data['Order Year'] = data['Order Date'].dt.year
data['Order Day of Week'] = data['Order Date'].dt.dayofweek

#Order Month: Order date se month extract karte hain.
#Order Year: Order date se year extract hota hai.
#Order Day of Week: Week ka day (0 for Monday, 6 for Sunday) extract kiya gaya hai.
```

```python
data.head()
```

# Monthly Sales Analysis

```python
# Mengelompokkan data berdasarkan bulan, lalu menjumlahkan total penjualan (Sales) per bulan
sales_by_month = data.groupby('Order Month')['Sales'].sum().reset_index()
# Membuat grafik garis (line chart) menggunakan Plotly Express
fig = px.line(sales_by_month,
              x='Order Month',
              y='Sales',
              title='Monthly Sales Analysis')
fig.show()
```

# Sales Analysis by Category

```python
# Mengelompokkan total sales berdasarkan kategori produk
sales_by_category = data.groupby('Category')['Sales'].sum().reset_index()

# Membuat donut chart untuk melihat proporsi sales per kategori
fig = px.pie(sales_by_category,
             values='Sales',               # Nilai yang divisualisasikan
             names='Category',             # Label untuk tiap potongan chart
             hole=0.5,                     # Ukuran lubang di tengah (donut chart)
             color_discrete_sequence=px.colors.qualitative.Pastel)  # Warna pastel

# Menampilkan persentase + label di dalam chart
fig.update_traces(textposition='inside', textinfo='percent+label')

# Menambahkan judul dan mengatur ukuran font judul
fig.update_layout(title_text='Sales Analysis by Category',
                  title_font=dict(size=24))

# Menampilkan grafik interaktif
fig.show()
```

# Sales Analysis by Sub-Category

```python
# Mengelompokkan total sales berdasarkan sub-kategori produk
sales_by_subcategory = data.groupby('Sub-Category')['Sales'].sum().reset_index()

# Membuat bar chart untuk membandingkan total penjualan per sub-kategori
fig = px.bar(sales_by_subcategory,
             x='Sub-Category',      # Sumbu X: nama sub-kategori
             y='Sales',             # Sumbu Y: total penjualan
             title='Sales Analysis by Sub-Category')  # Judul grafik

# Menampilkan grafik interaktif
fig.show()
```

# Monthly Profit Analysis

```python
# Mengelompokkan total profit berdasarkan bulan pemesanan
profit_by_month = data.groupby('Order Month')['Profit'].sum().reset_index()

# Membuat line chart untuk melihat tren profit bulanan
fig = px.line(profit_by_month,
              x='Order Month',   # Sumbu X: bulan
              y='Profit',        # Sumbu Y: total profit
              title='Monthly Profit Analysis')  # Judul grafik

# Menampilkan grafik interaktif
fig.show()
```

# Profit Analysis by Category

```python
# Mengelompokkan total profit berdasarkan kategori produk
profit_by_category = data.groupby('Category')['Profit'].sum().reset_index()

# Membuat donut chart untuk melihat proporsi profit per kategori
fig = px.pie(profit_by_category,
             values='Profit',               # Nilai profit yang akan divisualisasikan
             names='Category',              # Label tiap potongan donut
             hole=0.5,                      # Membuat chart jadi donut (ada lubang tengah)
             color_discrete_sequence=px.colors.qualitative.Pastel)  # Warna pastel

# Menampilkan persentase dan label di dalam chart
fig.update_traces(textposition='inside', textinfo='percent+label')

# Menambahkan judul grafik dan mengatur ukuran font
fig.update_layout(title_text='Profit Analysis by Category',
                  title_font=dict(size=24))

# Menampilkan grafik interaktif
fig.show()
```

# Profit Analysis by Sub-Category

```python
# Hitung total profit per sub-kategori
profit_by_subcategory = data.groupby('Sub-Category')['Profit'].sum().reset_index()

# Buat grafik batang dengan sumbu X = Sub-Category dan sumbu Y = total Profit
fig = px.bar(profit_by_subcategory, x='Sub-Category',
             y='Profit',
             title='Profit Analysis by Sub-Category')

# Tampilkan grafik
fig.show()

```

# Sales and Profit Analysis by Customer Segment

```python
# Hitung total Sales dan Profit per Segment
sales_profit_by_segment = data.groupby('Segment').agg({'Sales': 'sum', 'Profit': 'sum'}).reset_index()

# Pilih palet warna pastel untuk grafik
color_palette = colors.qualitative.Pastel

fig = go.Figure()

# Tambahkan bar untuk Sales per Segment
fig.add_trace(go.Bar(x=sales_profit_by_segment['Segment'],
                     y=sales_profit_by_segment['Sales'],
                     name='Sales',
                     marker_color=color_palette[0]))

# Tambahkan bar untuk Profit per Segment
fig.add_trace(go.Bar(x=sales_profit_by_segment['Segment'],
                     y=sales_profit_by_segment['Profit'],
                     name='Profit',
                     marker_color=color_palette[1]))

# Atur judul dan label sumbu grafik
fig.update_layout(title='Sales and Profit Analysis by Customer Segment',
                  xaxis_title='Customer Segment', yaxis_title='Amount')

# Tampilkan grafik
fig.show()
```

# Analyse sales-to-profit ratio

```python
# Hitung total Sales dan Profit per Segment
sales_profit_by_segment = data.groupby('Segment').agg({'Sales': 'sum', 'Profit': 'sum'}).reset_index()

# Hitung rasio Sales terhadap Profit untuk setiap Segment
sales_profit_by_segment['Sales_to_Profit_Ratio'] = sales_profit_by_segment['Sales'] / sales_profit_by_segment['Profit']

# Tampilkan kolom Segment dan Sales_to_Profit_Ratio
print(sales_profit_by_segment[['Segment', 'Sales_to_Profit_Ratio']])
```
